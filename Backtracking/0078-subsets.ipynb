# <center> 78. Subsets </center>


## Problem Description
[Click here](https://leetcode.com/problems/subsets/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The total number of subsets for an array of size n is $2^n$. 

We can make two choices for an element: 
- include it in the subset and move to the next 
- exclude it and move...

We need to explore all possible subsets. Use backtracking to explore all possible subsets.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store all subsets
- define a helper function for backtracking, it will consider two paths at each step: including the current index element and excluding it. In this way, it will generate all subsets <br>
*backtrack(current index, current path/subset)*
    - if the index is out of bounds, it means we have considered all elements
        - add the current subset copy to the result list and return
            - *(add the copy to avoid overwriting during future backtracking because the path is mutable and will be modified during recursion)*
    - add the current index element to the current subset and backtrack from next index i.e i + 1 to explore subsets with the current element
      - *(an element can appear only once in a subset, so we backtrack from the next index)*
    - remove the current index element from the current subset and backtrack from the next index i.e i + 1 to explore subsets without the current element
- start backtracking from index 0 with an empty subset
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(total recursive calls * work per call) → O(branching factor ^ recursion depth * copy cost) → O($2^n$ * n) → O(n * $2^n$)
  - *for each element, we have two choices: include or exclude, so the branching factor is 2*
  - *the recursion depth is n because we process one element per level*
  - *copying a subset takes O(n) time in the worst case*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + current subset + output list) → O(recursion depth + n + total subsets * subset size) → O(n + n + $2^n$ * n) → O(n * $2^n$)
  - *the current subset stores at most n elements*
  - *an array of size n has $2^n$ subsets*


## Code

```python
class Solution:

    def subsets(self, nums: List[int]) -> List[List[int]]:
        res = []

        def backtrack(i: int, path: List[int]) -> None:
            if i >= len(nums):
                res.append(path.copy())
                return
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
            backtrack(i + 1, path)

        backtrack(0, [])
        return res
```
