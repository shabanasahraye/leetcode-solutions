# <center> 131. Palindrome Partitioning </center>


## Problem Description
[Click here](https://leetcode.com/problems/palindrome-partitioning/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to split the string into all possible partitions such that every substring in the partition is a palindrome.

A string of length n has up to $2^{n - 1}$ possible partition patterns.

Since there are multiple ways to split the string, use backtracking to explore all possible partitions. For each possible substring, check whether it is a palindrome before adding it to the current partition.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a result list to store valid palindrome partitions
- define a helper function to check if a string is palindrome or not <br>
*is_palindrome(left index, right index)*
    - loop until left index < right index
        - if the left character is not same as the right character, the substring can't be a palindrome
            - return false
        - else characters are same, move left and right index pointers
    - if the loop completes without returning false, the substring is a palindrome, return true
- define a helper function for backtracking <br>
*backtrack(start position, current path/partition)*
    - if we have processed all characters (start == len(s)), the current partition is valid
        - add a copy of the current partition to the result list
            - *(add the copy to avoid overwriting during future backtracking because the path is mutable and will be modified during recursion)*
        - return
    - for each possible ending position of the next substring
        - if the substring ending with the current character is a palindrome
            - add it to the current path 
            - backtrack
            - remove the substring from the path
- start backtracking from index 0 as the start position
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(total recursive calls * work per call) → O(total states explored * work per call) → O(total partitions * work per partition) → O($2^n$ * n) → O(n * $2^n$)
  - *the branching factor varies because we can choose different substring lengths at each level*
  - *as the branching factor is variable, total recursive calls ≈ total states explored*
  - *there can be up to $2^n$ partitions*
  - *each partition may contain up to n characters, so palindrome checks, substring creation, and copying take O(n)*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + current path + output list) → O(recursion depth + current partition + total partitions * partition size) → O(n + n + $2^n$ * n) → O(n * $2^n$)
  - *recursion depth is at most n because we process at least one character per level*
  - *the current partition can contain at most n substrings*


## Code

```python
class Solution:

    def partition(self, s: str) -> List[List[str]]:
        res = []

        def is_palindrome(l: int, r: int) -> bool:
            while l < r:
                if s[l] != s[r]:
                    return False
                l += 1
                r -= 1
            return True

        def backtrack(start: int, path: List[str]) -> None:
            if start >= len(s):
                res.append(path.copy())
                return
            for i in range(start, len(s)):
                if is_palindrome(start, i):
                    path.append(s[start : i + 1])
                    backtrack(i + 1, path)
                    path.pop()

        backtrack(0, [])
        return res
```
