# <center> 51. N-Queens </center>


## Problem Description
[Click here](https://leetcode.com/problems/n-queens/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given n x n chessboard and n queens. We need to return the valid board configurations. A configuration is valid if the queens are placed in a way that they will not attack each other.

A queen can move in any direction i.e top, bottom, left, right, and across the diagonals. We need to place each queen in a different row, column, and diagonal so they will not be able to attack.

Use backtracking to explore all possible paths/configurations row by row and check if it is valid. A path is valid if there is no queen placed in the same row, column, or diagonal.

To speed up the process, don't check every cell. There are n queens and n rows and we need to place each in a row. So for every row, find a position (column) that will not result in a clash. For this track the occupied columns and diagonals. <br>
Use hash set to track/search, insert, and remove in O(1) time.

The left diagonal can be determined by (row - column). In the left diagonal, we move downwards, the row and column both increase by 1, and therefore the (row - column) remains constant.

The right diagonal can be determined by (row + column). In the right diagonal, we move upwards, the row decreases by 1, the column increases by 1, and therefore (row + column) remains constant.


## Approach
<!-- Describe your approach to solving the problem. -->
- create three sets to track the occupied columns, left diagonals, and right diagonals
- initialize a board of size n x n with dots to represent the empty cells
- create a list to store the valid board configurations
- define a helper function for backtracking <br>
*backtrack(row index)*
    - if current row index is equal to n, we have traversed all rows and placed all queens
        - the current configuration is valid, add it to the result list
            - *(convert the current board configuration (nxn matrix) to a list of strings i.e convert each row to a string and add all rows to a list)*
        - return
    - for each column in the current row
        - if there is a clash i.e current col or right diagonal or left diagonal is occupied
            - skip because we can't choose it to place the queen
        - else we can place a queen in the current cell
            - add current column, left and right diagonal to the sets to mark them as occupied
            - place queen in the current cell i.e mark current cell as 'Q'
            - backtrack to next row
            - unmark the column and diagonals from the sets
            - remove queen from the current cell
- start backtracking from the first row at index 0
- return the result


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(total recursive calls * work per call) → O(branching factor i.e choices at each level ^ recursion depth * constant work) → O(n ^ n * constant) → O($n^n$)
  - *at each level/row, we try placing a queen in up to n columns*
  - *recursion depth is n because we place one queen per row*
  - *each call does constant work excluding recursion*
    - *time for join() is ignored because it is done only for valid solutions, not at every recursive call*


<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + sets + board + output) → O(recursion depth + 3n + n² + all valid boards) → O(n + 3n + n² + total boards * board size) → O(n + 3n + n² + k * n²) → O(n² + k * n²)
  - *sets store at most n columns and diagonals*
  - *board uses n×n space*


## Code

```python
class Solution:

    def solveNQueens(self, n: int) -> List[List[str]]:
        cols = set()
        right_diag = set() 
        left_diag = set()
        board = [['.'] * n for _ in range(n)]
        res = []

        def backtrack(r: int) -> None:
            if r == n:
                res.append([''.join(row) for row in board])
                return
            for c in range(n):
                if c in cols or (r + c) in right_diag or (r - c) in left_diag:
                    continue
                cols.add(c)
                right_diag.add(r + c)
                left_diag.add(r - c)
                board[r][c] = 'Q'
                backtrack(r + 1)
                cols.remove(c)
                right_diag.remove(r + c)
                left_diag.remove(r - c)
                board[r][c] = '.'

        backtrack(0)
        return res
```
