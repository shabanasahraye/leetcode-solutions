## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Permutation I](https://leetcode.com/problems/permutations/description/), the difference is that Permutation II input contains duplicates.

A permutation is the rearrangement of the elements. The total number of permutations for n elements is n!

Use backtracking and swapping to explore all permutations and rearrange the elements. To handle duplicates, use a hash set or sort the input to avoid backtracking/recursion for them.


## Approach
<!-- Describe your approach to solving the problem. -->
*Use [Permutation I solution](https://leetcode.com/problems/permutations/solutions/3922226/on-n-time-on-n-space-solution-explained/) with check for the duplicates*

- set n = total number of elements
- create a result list to store the permutations
- define a helper function for backtracking <br>
*backtrack(start index)*
    - if current start index is same as n, we have considered all elements and current rearrangement/permutation is valid
        - add a copy of the current permutation to the result list and return
            - *(add the copy to avoid overwriting during future backtracking because path is a mutable list and will be modified during recursion)*
    - create a set to check for duplicates
    - for each index in the range start index to last index in the input list
        - if the element at the current index is not in the set, it is not a duplicate
            - add current index element to the set
            - swap the start index and current index element
            - backtrack
            - swap the start index and current index element again
- start backtracking from index 0 as the start position
- return result 


## Complexity
- Time complexity: O(total recursive calls * work per call) → O(recursion breadth * recursion depth * work per call) → O(total unique permutaions * n * constant) → O(n! * n * constant) → O(n * n!)
  - *we only generate unique permutations by skipping duplicates at each level*
  - *recursion depth is n because we fix one index per level*
  - *each call does constant work excluding recursion*
    - *time for copying is ignored because we only copy when we find a valid permutation, not at every recursive call*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + set + output list) → O(recursion depth + n + total permutations * permutation size) → O(n + n + n! * n) → O(n * n!)


## Code

```python
class Solution:

    def permuteUnique(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        res = []

        def backtrack(start: int) -> None:
            if start == n:
                res.append(nums.copy()) 
            seen = set() 
            for i in range(start, n):
                if nums[i] not in seen:
                    seen.add(nums[i])
                    nums[start], nums[i] = nums[i], nums[start]
                    backtrack(start + 1)
                    nums[start], nums[i] = nums[i], nums[start]
        
        backtrack(0)
        return res
```
