## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Permutation I](https://leetcode.com/problems/permutations/description/), the difference is that Permutation II input contains duplicates.

A permutation is a rearrangement of the elements. <br>
For n elements, there are n! possible permutations.

Use backtracking to explore all possible arrangements and generate every permutation.

To generate all permutations, fix one position at a time and try every remaining element at that position.

Use swapping to fix/place an element at the current position. Recursively generate permutations for the remaining positions, then swap back to try other choices.

To handle duplicates, use a hash set at each recursion level or sort the input to avoid generating duplicate permutations.


## Approach
<!-- Describe your approach to solving the problem. -->
*Use [Permutation I solution](https://leetcode.com/problems/permutations/solutions/3922226/on-n-time-on-n-space-solution-explained/) with check for the duplicates*

- set n = total number of elements
- create a result list to store the permutations
- define a helper function for backtracking <br>
*backtrack(start index)*
    - if current start index is same as n, we have considered all elements and current rearrangement/permutation is valid
        - add a copy of the current permutation to the result list and return
            - *(add the copy to avoid overwriting during future backtracking because path is a mutable list and will be modified during recursion)*
    - create a set to check for duplicates
    - for each index in the range start index to n in the input list
        - if the element at the current index is not in the set, it is not a duplicate
            - add current index element to the set
            - swap the start index and current index element
              - *(this is to fix the current element at the current position)*
            - backtrack to explore permutations for the remaining positions
            - swap the start index and current index element again
              - *(this is to restore the original arrangement and try the next element)*
- start backtracking from index 0 as the start position
- return result 


## Complexity
- Time complexity: O(backtracking) → O(total recursive calls * work per call) → O(total states explored * work per call) → O(total unique permutations * copy cost) → O(n! * n) → O(n * n!)
  - *branching factor is variable because the number of available choices decreases at each level*
  - *as the branching factor is variable, total recursive calls ≈ total states explored*
  - *each valid permutation contains n elements and copying it costs O(n)*

- Space complexity: O(recursion stack + set + output list) → O(recursion depth + set + total unique permutations * permutation size) → O(n + n + n! * n) → O(n * n!)
  - *recursion depth is n because we fix one position per level*
  - *set stores at most n elements per level*
  - *output stores all unique permutations, each of size n*


## Code

```python
class Solution:

    def permuteUnique(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        res = []

        def backtrack(start: int) -> None:
            if start == n:
                res.append(nums.copy()) 
                return
            seen = set() 
            for i in range(start, n):
                if nums[i] not in seen:
                    seen.add(nums[i])
                    nums[start], nums[i] = nums[i], nums[start]
                    backtrack(start + 1)
                    nums[start], nums[i] = nums[i], nums[start]
        
        backtrack(0)
        return res
```
