## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Combination Sum I](https://leetcode.com/problems/combination-sum/), the difference is that combination sum II input contains duplicates and the result must not contain duplicate combinations.

We need to find all valid and unique combinations. A combination is valid if it contains unique elements and the sum of the elements is equal to the target.

To avoid duplicate combinations, we can make two choices for an element:
- include the element in the combination and move ahead because a number can only appear once in a combination
- don't include the element and move ahead

Use bactracking to explore all combinations and check if it is valid. As input contains duplicates, sort the input first to group duplicates and avoid recursion for them.

## Approach
<!-- Describe your approach to solving the problem. -->
*Same as the [Combinations Sum I solution](https://leetcode.com/problems/combination-sum/solutions/3917442/o-2-n-time-o-n-space-solution-explained/), only sort the input to handle duplicates and backtrack from the next index so that each number appears only once in a combination.*
- sort the input list
- create a result list to store the valid combinations
- define a helper function for backtracking <br>
*backtrack(current index, current combination sum, current path/combination)*
    - if current combination sum is equal to the target, it is a valid combination
        - add a copy of the current combination to the result list and return
            - *(add the copy to avoid overwriting this combination during future backtracking because path is a mutable list and will be modified during recursion)*
    - if the index is out of bounds or current combination sum is not valid, return
    - add the current index element to the current combination and backtrack from next index i.e i + 1 to explore combinations with the current element
        - *(a number can only appear once in a combination, so we backtrack from the next index)*
    - remove the current index element from the current combination
    - increment the index while there are duplicates i.e current and next element is the same
    - backtrack from the next index i.e i + 1 to explore combinations without the current element
- start backtracking from index 0
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(sort + total recursive calls * work per call) → O(nlogn + total choices at each level ^ recursion depth * work per call) → O(nlo n + branching at each step depends on remaining sum and candidate choices ^ variable recursion depth * constant) → O(nlogn + branching factor ^ recursion depth) → O(nlogn + exponential) → O(exponential)
  - *even though at each level we either pick or skip the element, the complexity is exponential not $2^n$ because:*
    - *recursion depth is variable and depends on target sum, we are not making n fixed decisions*
    - *recursion tree structure is irregular due to target-based pruning and duplicate skipping: it is not a perfect binary tree, one branch can go very deep while others are pruned or skipped*
  - *each call does constant work excluding recursion*
    - *time for copying is ignored because we only copy when we find a valid combination, not at every recursive call*
    - *same goes for the while loop used to skip duplicates because it does not run in every call (amortized)*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(sort + recursion stack + output list) → O(n + recursion depth + total combinations * combination size) → O(n + variable recursion depth + total valid combinations * combination size)
  - *number of valid combinations depends on input and target*
  - *each combination has variable length based on sum to target*


## Code

```python
class Solution:

    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort() 
        res = []

        def backtrack(i: int, total: int, path: List[int]) -> None:
            if total == target:
                res.append(path.copy())
                return 
            if i >= len(candidates) or total > target:
                return
            path.append(candidates[i])
            backtrack(i + 1, total + candidates[i], path)
            path.pop()
            while i + 1 < len(candidates) and candidates[i] == candidates[i + 1]:
                i = i + 1
            backtrack(i + 1, total, path)

        backtrack(0, 0, [])
        return res
```
