## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Combination Sum I](https://leetcode.com/problems/combination-sum/), the difference is that combination sum II input contains duplicates and a number can appear only once in the combination.

We need to find all valid and unique combinations. A combination is valid if it contains unique elements and the sum of the elements is equal to the target.

To generate unique combinations with unique elements, we can make two choices for an element:
- include the element in the combination and move to the next index so it can't be reused
- don't include the element and move to the next index

Use bactracking to explore all combinations and check if they are valid. <br>
As input contains duplicates, sort the input first to group duplicates and avoid generating duplicate combinations.

## Approach
<!-- Describe your approach to solving the problem. -->
*Same as the [Combinations Sum I solution](https://leetcode.com/problems/combination-sum/solutions/3917442/o-2-n-time-o-n-space-solution-explained/), only sort the input to handle duplicates and backtrack from the next index so that each number appears only once in a combination.*
- sort the input list
- create a result list to store the valid combinations
- define a helper function for backtracking <br>
*backtrack(current index, current combination sum, current path/combination)*
    - if current combination sum is equal to the target, it is a valid combination
        - add a copy of the current combination to the result list and return
            - *(add the copy to avoid overwriting during future backtracking because path is a mutable list and will be modified during recursion)*
    - if the index is out of bounds or current combination sum is not valid, return
    - add the current index element to the current combination and backtrack from next index i.e i + 1 to explore combinations with the current element
        - *(a number can only appear once in a combination, so we backtrack from the next index)*
    - remove the current index element from the current combination
    - increment the index while there are duplicates i.e current and next element is the same
    - backtrack from the next index i.e i + 1 to explore combinations without the current element
- start backtracking from index 0
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(sort + backtracking) → O(sort + total recursive calls * work per call) → O(nlogn + branching factor ^ recursion depth * constant) → O(nlogn + $2^n$) → O($2^n$)
  - *branching factor is 2 because for each element we have two choices: include or exclude*
  - *as the branching factor is constant, total recursive calls =  branching factor ^ recursion depth*
  - *recursion depth is at most n because each element can be used at most once per path*
    - *unlike Combination Sum I, the depth depends on the number of elements (n), not the target, because elements cannot be reused*
  - *duplicate skipping removes redundant branches but does not change the worst-case complexity*
  - *each call does O(1) work excluding recursion*
    - *copying a valid combination costs O(k), but only when a solution is found (at leaf states)*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(sort + recursion stack + current path + output list) → O(sort + recursion depth + current combination + total valid combinations * combination size) → O(n + n + n + total valid combinations * combination size) → O(n + output)
  - *current path stores at most n elements*
  - *output list stores all valid unique combinations, each of variable length*


## Code

```python
class Solution:

    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort() 
        res = []

        def backtrack(i: int, total: int, path: List[int]) -> None:
            if total == target:
                res.append(path.copy())
                return 
            if i >= len(candidates) or total > target:
                return
            path.append(candidates[i])
            backtrack(i + 1, total + candidates[i], path)
            path.pop()
            while i + 1 < len(candidates) and candidates[i] == candidates[i + 1]:
                i = i + 1
            backtrack(i + 1, total, path)

        backtrack(0, 0, [])
        return res
```
