# <center> 1239. Maximum Length of a Concatenated String with Unique Characters </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-length-of-a-concatenated-string-with-unique-characters/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a list of strings. We need to find the max-length valid subsequence. <br>
A subsequence is valid if it has no duplicate letters.

For each string, we have two choices:
- include it in the current subsequence
- exclude it and move ahead

Use backtracking to explore all possible subsequences and check if they are valid. 

To optimize:
- remove duplicate strings and strings with duplicate letters
- use hashset to search in O(1) time


## Approach
<!-- Describe your approach to solving the problem. -->
- preprocess the input i.e remove the duplicate strings and strings with duplicate letters
  - *there are no duplicate strings in any of the test cases, so we will only handle the duplicate letter strings*
- if after preprocessing:
    - there are no strings left, return 0
    - only 1 string left, return its length
- use a set to store current subseq letters and avoid duplicates
- define a helper function to check if there's overlap i.e adding current string to current subseq will result in duplicate letters <br>
overlap(current subseq, current string)
    - for each character in the string
        - if the character is in the subseq
          - return true because there is overlap
    - if the loop completes without returning true, there is no overlap, return false
- define a helper function for backtracking <br>
*backtrack(current string index)*
    - if the current index exceeds the last, we have considered all strings
      - return the current subseq length
    - set res = 0 to track the max length valid subseq
    - if adding the current string to the current subseq will not result in overlap
        - add the current string characters to the current subseq
        - backtrack to explore other subsequence and store the result in res
        - remove the current string characters from the current subseq
    - else exclude the current string
      - move to next index and backtrack
    - max(include, exclude)
- start backtracking from the first string at index 0 and return the result
     

## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(branching factor ^ recursion depth) → O($2^n$)
  - *for each string, we have two choices: include it or exclude it*
  - *n = number of strings after preprocessing*
  - *character overlap check takes O(26) because string contains lowercase English letters*
    - *no need to count the duplicates because we remove the strings with duplicate letters during preprocessing*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + current path) → O(recursion depth + 26) → O(n + 26) → O(n)
  - *recursion depth is at most n becuase we take one decision per string*
  - *current path stores at most 26 unique characters*


## Code

```python
class Solution:

    def maxLength(self, arr: List[str]) -> int:
        arr = [i for i in arr if len(i) == len(set(i))]
        if not arr:
            return 0
        elif len(arr) == 1:
            return len(arr[0])
        path = set()

        def overlap(path, s):
            for c in s:
                if c in path:
                    return True
            return False

        def backtrack(i):
            if i == len(arr):
                return len(path)
            res = 0
            if not overlap(path, arr[i]):
                for c in arr[i]:
                    path.add(c)
                res = backtrack(i + 1)
                for c in arr[i]:
                    path.remove(c)
            return max(res, backtrack(i + 1)) 

        return backtrack(0)
```
