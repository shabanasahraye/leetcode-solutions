# <center> 77. Combinations </center>


## Problem Description
[Click here](https://leetcode.com/problems/combinations/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need all combinations of size k chosen from the range 1 to n. 

The total number of combinations of size k chosen from n elements is $n^k$.

Use backtracking to explore all possible combinations

For each element, we can make two choices:
- include it in the combination and move to the next because a number can appear only once in a combination
- exclude it and move...


## Approach
<!-- Describe your approach to solving the problem. -->
- create a result list to store the combinations
- define a helper function for backtracking <br>
*backtrack(current start position, current path/combination)*
    - if there are k elements in the current combination
        - add a copy of the current combination to the result list and return
           - *(add the copy to avoid overwriting during future backtracking because the path is mutable and will be modified during recursion)*
    - for each element in range start position to end of the input range n
        - add the current element to the combination 
        - backtrack to explore combinations with the current element
            - *(a number can appear only once in a combination, so backtrack from the next index)*
        - remove the current element to explore combinations without the current element
- start backtracking from 1 as the start position
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(total recursive calls * work per call) → O(total combinations * copy cost) → O($n^k$ * k) → O(k * $n^k$)
  - *each valid combination contains k elements, so copying takes k time*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + current path + output list) → O(recursion depth + k + total combinations * combination size) → O(k + k + $n^k$ * k) → O($n^k$ * k) → O(k * $n^k$)
  - *recursion depth is at most k because we choose one element per level*
  - *current path stores at most k elements*


## Code

```python
class Solution:

    def combine(self, n: int, k: int) -> List[List[int]]:
        res = []

        def backtrack(start: int, path: List[int]) -> None:
            if len(path) == k:
                res.append(path.copy())
                return
            for i in range(start, n + 1):
                path.append(i)
                backtrack(i + 1, path)
                path.pop()

        backtrack(1, [])
        return res
```
