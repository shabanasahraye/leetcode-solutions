## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given matchsticks and we need to find if we can make a square using all the matchsticks.

A square has 4 sides of equal length. Find the length of a single side and use backtracking to assign each matchstick to one of the four sides while ensuring that no side exceeds the required length.

single side length = perimeter / 4 = sum of all matchsticks / 4

Sort the matchsticks in decreasing order so that bigger matchsticks will get processed earlier and will lead to a negative conclusion first if there is no solution.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there are less than 4 matchsticks available
        - return false becuase we can't make a square
    - if the perimeter is not divisible by 4 or the longest matchstick is greater than the length of a side
        - return false becuase we can't make a square
- sort the matchsticks in decreasing order
- initialize an array of size 4 to track the current length of the 4 sides 
- use a helper function to backtrack <br>
*backtrack(current index)*
    - if the current index exceeds the last, we have assigned all matchsticks to a side and can form a square
        - return True
    - for each of the 4 sides
        - if current matchstick is valid i.e adding the current index matchstick to the current side results in a length less than or equal to the actual required length
            - add the current matchstick length to the side 
            - backtrack to check if we can add any more matchsticks to the current side i.e explore all combinations including the current matchstick, return true if we can
            - remove the current matchstick from the side to explore placing it on another side
            - if the current side becomes 0 after removal
                - break to avoid duplicate work on other empty sides because all empty sides are equivalent
    - if the loop completes without returning true, there is no combination of matchsticks that can sum up to the side of a square, return false
- start backtracking from the first matchstick at index 0 and return the result


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(sort + backtracking) → O(sort + total recursive calls * work per call) → O(sort + branching factor ^ recursion depth * work per call) → O(n log n + $4^n$ * constant) → O($4^n$)
  - *branching factor is 4 because each matchstick can go to any of the 4 sides*
  - *as the branching factor is constant, total recursive calls = branching factor ^ recursion depth*
  - *recursion depth is n because we process one matchstick per level*
  - *work per call is constant because we only update/check one of the 4 sides*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(sort + recursion stack + sides array) → O(sort + recursion depth + 4) → O(n + n + 4) → O(n)


## Code

```python
class Solution:

    def makesquare(self, matchsticks: List[int]) -> bool:
        if len(matchsticks) < 4:
            return False
        perimeter = sum(matchsticks)
        length = perimeter // 4
        if perimeter % 4 or max(matchsticks) > length:
            return False
        matchsticks.sort(reverse=True)
        sides = [0] * 4

        def backtrack(i: int) -> bool:
            if i == len(matchsticks):
                return True
            for j in range(4):
                if sides[j] + matchsticks[i] <= length:
                    sides[j] += matchsticks[i]
                    if backtrack(i + 1):
                        return True
                    sides[j] -= matchsticks[i]
                    if sides[j] == 0:
                        break
            return False

        return backtrack(0)
```
