# <center> 90. Subsets II </center>


## Problem Description
[Click here](https://leetcode.com/problems/subsets-ii/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Subset I](https://leetcode.com/problems/subsets/), the difference is that Subset II input contains duplicates.

The total number of subsets for an array of size n is $2^n$. 

We can make two choices for an element: 
- include it in the subset and move to the next
- exclude it and move...

Use backtracking to explore all possible subsets. <br>
Sort the input first to group the duplicates and avoid generating duplicate subsets.


## Approach
<!-- Describe your approach to solving the problem. -->
*same as [Subset I solution](https://leetcode.com/problems/subsets/solutions/3914626/o-n-2-n-time-o-n-2-n-space-solution-explained/), just sort the input first to handle the duplicates*
- set res = a list to store the subsets
- sort the input
- define a helper function for backtracking <br>
*backtrack(current index, current path/subset)*
    - if the index is out of bounds, we have considered all elements
        - add the current subset copy to the result list, and return
            - *(add the copy to avoid overwriting during future backtracking because the path is mutable and will be modified during recursion)*
    - add the current index element to the current subset
    - backtrack from next index i.e i + 1 to explore subsets with the current element
        - *(an element can appear only once in a subset, so we backtrack from the next index)*
    - remove the current index element from the current subset
    - increment the index while there are duplicates i.e current and next elements are the same
    - backtrack from next index i.e i + 1 to explore subsets without the current element
- start backtracking from index 0 with an empty subset
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(sort + backtracking) → O(sort + total recursive calls * work per call) → O(sort + branching factor ^ recursion depth * copy cost) → O(nlogn + $2^n$ * n) → O(n * $2^n$)
  - *branching factor is 2 because for each element, we have two choices: include or exclude*
  - *as the branching factor is constant, total recursive calls = branching factor ^ recursion depth*
  - *recursion depth is n because we process one element per level*
  - *copying takes O(n) because each subset can contain up to n elements*
  - *while-loop to skip duplicates reduces the recurscive calls but does not change the overall complexity*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(sort + recursion stack + current subset + output list) → O(sort + recursion depth + current subset + total subsets * subset size) → O(n + n + n + $2^n$ * n) → O(n * $2^n$)
  - *the current subset stores at most n elements*
  - *an array of size n has $2^n$ subsets*


## Code

```python
class Solution:

    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort()

        def backtrack(i: int, path: List[int]) -> None:
            if i == len(nums):
                res.append(path.copy())
                return
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
            while i + 1 < len(nums) and nums[i] == nums[i + 1]:
                i += 1
            backtrack(i + 1, path)

        backtrack(0, [])
        return res
```
