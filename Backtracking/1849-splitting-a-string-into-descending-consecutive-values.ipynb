# <center> 1849. Splitting a String Into Descending Consecutive Values </center>


## Problem Description
[Click here](https://leetcode.com/problems/splitting-a-string-into-descending-consecutive-values/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a string of digits and need to split it into substrings such that the substrings are consecutive numbers in descending order (difference = 1).

total number of ways to split a string = $2^n$

Use backtracking to explore all possible splits points and check if they are valid. <br>
A split is valid if the resulting substring values are in descending order and the difference is 1 (substr2 - substr1 = 1).


## Approach
<!-- Describe your approach to solving the problem. -->
- define a helper function for backtracking <br>
*backtrack(current index, current partitions, previous substring value)*
    - if the current index exceeds the last, we have reached the end of the string and have considered all digits
        - check if we were able to split the string i.e current partitions >= 2 and return the result
    - loop in range current index i to string s size to check for the next digit <br>
    *for each possible substring starting at index i*
        - convert current substring to a number
        - if this is the first number or the difference between the previous and current is 1, current split is valid
            - backtrack from next index to explore whether we can make another split and return true if we can
    - if the loop completes without returning true, we can't split the string, return false
- start backtracking from the first digit at index 0 and return the result


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(branching factor ^ recursion depth) → O($2^n$)
  - *at each index, we try multiple split points*
  - *at each position (gap between tow chars), we have two choices: cut or not cut*
  - *recursion depth is at most n due to split positions*
  - *average time is much smaller due to pruning when a valid descending condition fails early*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack) → O(recursion depth) → O(n)


## Code

```python
class Solution:

    def splitString(self, s: str) -> bool:
    
        def backtrack(i: int, k: int, prev: int) -> bool:
            if i == len(s):
                return k >= 2
            for j in range(i, len(s)):
                cur = int(s[i:j + 1])
                if prev == math.inf or cur == prev - 1:
                    if backtrack(j + 1, k + 1, cur):
                        return True
            return False
        
        return backtrack(0, 0, math.inf)
```
