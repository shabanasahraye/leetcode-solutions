# <center> 93. Restore IP Addresses </center>


## Problem Description
[Click here](https://leetcode.com/problems/restore-ip-addresses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a string of digits and we need to return all valid IP address combinations.

A valid IP address consists of 4 blocks. Each block: 
- is separated by a dot
- can contain up to 3 digits
- represents a number between 0 and 255 without leading zeros

Use backtracking to explore all possible combinations and check if they are valid.

## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the input string size is more than 12, there can't be a valid combination because an ip address can't have more than 12 digits
        - return an empty list
- create a list to store valid combinations
- define a helper function for backtracking <br>
*backtrack(start position i.e current digit index, current path/combination)*
    - if the current combination contains 4 blocks and we have used all digits (start == len(s)), the combination is valid
        - convert the combination into an IP address and add it to the result
        - return
    - elif we already have 4 blocks or we have used all digits, the combination is not valid
        - return
    - for each possible block length from 1 to 3 digits
        - if the current position is out of bound i.e current block length > string
            - return
        - if the block contains more than one digit and starts with 0
            - return because leading zeros are not allowed
        - get the number (substring) representing the current block
        - if the number is greater than 255:
            - return because increasing the block length will only make the number larger
        - else backtrack to explore and generate other blocks
- start backtracking from the first digit at index 0


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(branching factor ^ recursion depth) → O($3^4$) → O(81) → O(1)
  - *the recursion depth is 4 because a valid IP address contains 4 blocks*
  - *at each level, we can choose a block length of 1, 2, or 3 digits, giving a branching factor of 3*
  - *ignore time for join() because it is not called in each recursion and its time is 4 because it is joining 4 blocks*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + output list) → O(recursion depth + 19) → O(4 + 19) → O(1)
  - *the output list stores at most 19 valid IP addresses, which is the maximum possible for any input*


## Code

```python
class Solution:

    def restoreIpAddresses(self, s: str) -> List[str]:
        if len(s) > 12:
            return []
        res = []

        def backtrack(start: int, path: List[str]) -> None:
            if len(path) == 4 and start == len(s):
                res.append('.'.join(path))
                return
            if len(path) == 4 or start == len(s):
                return
            for length in range(1, 4):
                if start + length > len(s):
                    return 
                if length > 1 and s[start] == '0':
                    return 
                num = s[start: start + length]
                if int(num) > 255:
                    return
                backtrack(start + length, path + [num])

        backtrack(0, [])
        return res
```
