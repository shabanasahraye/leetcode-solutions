## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given some binary strings representing numbers in the range 0 to n. We need to return any of the missing binary number.

Approaches:
1. **Backtracking** <br>
Use backtracking to generate all binary strings of length n. At each index, try '0' and '1'. Check if the generated string is present in the input or not.
Convert the input to hash set to search in O(1) time.

2. **Binary Enumeration** <br>
Traverse all decimal numbers in the range 0 to n. Find the binary and check if it is present in the input list. <br>
You can also first convert all numbers to binary strings and then check but with that you would be doing some unnecessary conversions. <br>
Convert the input to hash set to search in O(1) time.

3. **Cantor's Diagonalization** <br>
According to this, a binary number missing from a binary sequence can be constructed by flipping the 1st bit of str1, flipping the 2nd bit of str2, ... and flipping the nth bit of strn. <br>
For example:  <br>
input strings = ['111', '001', '010'] i.e [7, 1, 2] <br>
missing numbers from the sequence are 3, 4, 5 <br>
we need only one missing number and the first missing number can be formed as: <br>
    - get 1st bit of string1 + get 2nd bit of string2 + get 3rd bit of string3 <br>
      1 + 0 + 0 <br>
    - flip the bits <br>
      0 + 1 + 1 <br>
    - concatenate <br>
      0 + 1 + 1 = 011 = 3  <br>
 

## Approach
<!-- Describe your approach to solving the problem. -->
**Backtracking Approach**
- convert the input list to a hash set for fast lookup
- define a helper function for backtracking <br>
*backtrack(current index, current path i.e binary string in list form)*
  - if the current string length == n
    - join path to get the current string
    - if the string is not in the input set
        - return the current string
    - else return none
  - *try both choices (0, 1) at each position to build all possible strings*
    - *try '0'*
      - add '0' and move to next index
      - backtrack
      - if result is found, return it
      - remove '0' from the path (backtrack step)
    - *try '1'*
      - add '1' and move to next index
      - backtrack
      - if result is found, return it
      - remove '1' from the path (backtrack step)
  - if no valid string is found, return None
- start backtracking from index 0 with 0 as starting string and  return the result

**Binary Approach**
- convert the input strings list to a set to search in O(1)
- set n = size of the input list
- loop in range 0 to n + 1 to check all numbers
    - get the binary string of the current number
        - *(binary string is prefixed with 0b, therefore formatting is required)*
    - if the current number binary string is not in the set, the current number is a missing number
        - return its binary string

**Cantor Approach**
- set res = a list to store the missing binary number bits
    - *(list because string concatenation is expensive)*
- loop in range 0 to the input list size to get the required bit
    - get the required bit i.e 1st bit of 1st string, 2nd bit of 2nd string etc, flip it
    - add it to the result list
- convert the result list to string to get the missing binary number and return it


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity:
    - Backtracking Approach: O(backtracking) → O(branching factor ^ recursion depth) → O($2^n$)
        - *at each position we try two choices: '0' or '1'*
        - *recursion depth is n becuause*
    - Binary Approach: O(set conversion + array traversal * decimal to binary conversion) → O(n + n * log(number)) → O(n + n) → O(n)
    - Cantor Approach: O(array traversal + converting list to string) → O(n + size of a binary string) → O(n + size of binary of 16 because n <= 16) → O(n + 5) → O(n)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity:
    - Backtracking Approach: O(recursion stack) → O(recursion depth) → O(n)
    - Binary Approach: O(set + decimal to binary conversion) → O(n + log(number)) → O(n)
    - Cantor Approach: O(result list) → O(n)
