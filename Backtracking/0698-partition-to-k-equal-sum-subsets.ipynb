# <center> 698. Partition to K Equal Sum Subsets </center>


## Problem Description
[Click here](https://leetcode.com/problems/partition-to-k-equal-sum-subsets/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find if it is possible to divide the array into k subsets of equal sum.

The sum of each subset should be equal to  sum(array) / k

For each element, we can make two choices:
- include it in the subset
- exclude it

Use backtracking to explore all possible subsets and check if it is a valid subset i.e sum(subset) = sum(array) / k <br>
To optimize, sort the array and use a hash set.


## Approach
<!-- Describe your approach to solving the problem. -->
- find the total sum of the array
- check base cases 
    - if the total sum is not divisible by k, it is not possible to have k subsets of equal sum, return false
- sort the array in descending order
- find the target sum for each subset
  - *target = total // k*
- create a hashset to check for duplicates
    - *(we will add indices to the hashset, if index is in set, it means we have used that index element)*
- define a helper function for backtracking <br>
*backtrack(current index, remaining subsets, current subset sum)*
    - if we have found k subsets
      - return true
    - if current subset sum is equal to the target sum, we have found a subset
        - reduce k and backtrack to start building next subset from index 0
        - return result
    - loop through array starting from index i to check for the remaining elements
        - if current element is a duplicate and previous was not used
            - skip
        - if the current element is already used i.e its index is in the hashset or adding it to the current subset exceeds the target sum
            - skip
        - mark index as used i.e to include it in the current subset
        - backtrack to find the k subsets, return true if it is possible
        - unmark index (backtrack)
    - if the loop executes without returning true, it is not possible to divide the array into k subsets, return false
- start backtracking from the first element at index 0


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(sort + backtracking) → O(nlogn + total recursive calls * work per call) → O(nlogn + total subsets * n) → O(nlogn + $2^n$ * n) → O(n * $2^n$)
  - *each element has two choices: include or exclude*
  - *work per recursive call is n due to iteration / checks over remaining elements*
  - *k does not control recursion complexity, it only controls how many times we restart after completing a valid subset*
    - *all subsets are generated in a single recursion tree not k separate phases*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(sort + hashset + recursion stack) → O(n + n + recursion depth) → O(n + n + n) → O(n)
  - *hashset stores at most n indices*
  - *recursion depth is at most n because each element is processed once in the recursion path*
  - *current path stores at most k elements*

## Code

```python
class Solution:

    def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
        total = sum(nums)
        if total % k:
            return False
        nums.sort(reverse = True)
        target = total / k
        seen = set()

        def backtrack(i: int, k: int, cur: int) -> bool:
            if k == 0:
                return True
            if target == cur:
                return backtrack(0, k - 1, 0)
            for j in range(i, len(nums)):
                if j > 0 and (j - 1) not in seen and nums[j] == nums[j - 1]:
                    continue
                if j in seen or cur + nums[j] > target:
                    continue
                seen.add(j)
                if backtrack(j + 1, k, cur + nums[j]):
                    return True
                seen.remove(j)
            return False

        return backtrack(0, k, 0)
```
