# <center> 39. Combination Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/combination-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find all valid and unique combinations. A combination is valid if the sum of elements is equal to the target. <br>
A number can be used multiple times in the same combination.
 

To generate unique combinations, we can make two choices for an element: 
- include the element in the combination and stay at the same index so it can be reused
- don't include the element and move to the next index

Use backtracking to explore all combinations and check if they are valid.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the valid combinations
- define a helper function for backtracking <br>
*backtrack(current index, current combination sum, current path/ combination)*
    - if current combination sum is equal to the target, it is a valid combination
        - add a copy of current combination to the result list and return
            - *(add the copy to avoid overwriting during future backtracking because path is a mutable list and will be modified during recursion)*
    - if index is out of bounds i.e we have run out of elements to explore or if sum exceeds target i.e current combination is not valid
      - return
    - add the current index element to the combination and backtrack from the same index to explore combinations with the current element
        - *(a number can appear multiple times in a combination, so we backtrack from the same index)*
    - remove the current index element and backtrack from the next index to explore combinations without the current element
- start backtracking from index 0
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(backtracking) → O(total recursive calls * work per call) → O(branching factor ^ recursion depth * work per call) → O($2^{\frac{target}{\min(candidates)}}$ * constant) → O($2^{\frac{target}{\min(candidates)}}$)
  - *the branching factor is 2 because for each candidate we have two choices: include or exclude*
  - *as the branching factor is constant, total recursive calls = branching factor ^ recursion depth*
  - *the recursion depth depends on the target sum, not on n, because an element can be reused*
    - *in the worst case, we repeatedly choose the smallest candidate until the target is reached*
    - *maximum recursion depth ≈ target / min(candidates)*
  - *each call performs O(1) work excluding recursive calls*
    - *copying a valid combination costs O(k), but only when a solution is found (at leaf states)*
  
<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack + current path + output list) → O(recursion depth + current combination + total valid combinations * combination size) → O(target/min(candidates) + target/min(candidates) + output) → O(target/min(candidates) + output)
  - *the current combination can contain at most target / min(candidates) elements*
  - *output list stores all valid unique combinations, each of variable length*
  - *number of valid combinations depends on input and target*


## Code

```python
class Solution:

    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        res = []

        def backtrack(i: int, total: int, path: List[int]) -> None:
            if total == target:
                res.append(path.copy())
                return
            if i >= len(candidates) or total > target:
                return
            path.append(candidates[i])
            backtrack(i, total + candidates[i], path)
            path.pop()
            backtrack(i + 1, total, path)

        backtrack(0, 0, [])
        return res
```
