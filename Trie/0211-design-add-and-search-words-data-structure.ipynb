## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use Trie data structure.

The Trie data structure, also known as a prefix tree, allows efficient insertion and search of words character by character. It stores words in a tree, where each node represents a character in the word.

Since the search word may contain '.' which can match any letter, a Trie enables us to efficiently explore all possible paths only when needed instead of checking every stored word.


## Approach
<!-- Describe your approach to solving the problem. -->
**Class TrieNode**

**init()**
- set children = a hashmap to store the child nodes
    - *key = characters*
    - *value = child nodes*
- set end = false to track if a node is the last node (character) of  a word

**Class WordDictionary**

**init()**
- create a root node

**insert()** <br>
*start from the root node and create a new node for each character in the word if the node is not already there. Mark the last node i.e end of the word by setting the last node end attribute to true*
- set cur = root node to track current node
- for each character in the word
    - if the character is not a child node of the current node, add it as the child node
    - set the current node to the child node of the character to move to the next node
- at the end of the loop, the current node will point to the last node (character) of the word. So, set the end attribute to true

**search()** <br>
*searching for a word in a trie is like finding a path from the root node to a word-ending node using DFS. It is mentioned that a word may contain dots '.' where a dot can be matched with any letter. Therefore, we have to explore all possible paths.*
- define a helper function for DFS <br>
*dfs(current node, current character index in the word)*
    - if the current index equals the length of the word
        - check if the current node is the last node (character) of the word and return the result
    - get the character at the current index in the search word
    - if the character is not a '.'
        - if the character is a child node
          - search for the character in the child nodes
        - else return false
    - else the character is a '.'
      - search in all child nodes of the current node to find a match
- start DFS from the root node with index 0 and return the result


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity:
    - init(): O(1)
    - addWord(): O(word length) → O(n)
    - search(): O(backtracking DFS) → O(total recursive calls * work per call) → O(total states explored * work per call) → O($26^n$ * constant) → O($26^n$)
        - *branching factor = number of child nodes explored when the current character is '.'*
        - *branching factor is variable because different trie nodes can have different numbers of children (from 0 to 26 for lowercase English letters)*
        - *as the branching factor is variable, total recursive calls = total states explored*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity:
    - init(): O(1)
    - addWord(): O(word length) → O(n) 
        - *in the worst case, a new node is created for each character in the word*
    - search(): O(recursion stack) → O(recursion depth) → O(n)
        - *recursion depth is n because we process one character per level*

## Code

```python
class TrieNode:

    def __init__(self):
        self.children = defaultdict(TrieNode)
        self.last = False


class WordDictionary:

    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word: str) -> None:
        cur = self.root
        for c in word:
            if c not in cur.children:
                cur.children[c] = TrieNode()
            cur = cur.children[c]
        cur.last = True

    def search(self, word: str) -> bool:

        def dfs(node: TrieNode, index: int) -> bool:
            if index == len(word):
                return node.last
            c = word[index]
            if c != '.':
                if c in node.children:
                    return dfs(node.children[c], index + 1)
                return False
            return any(dfs(child, index + 1) for child in node.children.values())

        return dfs(self.root, 0)
        
    
        


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
