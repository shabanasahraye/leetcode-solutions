# <center> 22. Generate Parentheses </center>


## Problem Description
[Click here](https://leetcode.com/problems/generate-parentheses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use DFS/backtracking to generate all valid balanced parentheses combinations.

At each step:
- add ( if we still have opening brackets left
- add ) only if it keeps the string valid


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the result i.e valid combinations
- define a helper function for DFS
    - dfs(left brackets count, right brackets count, current string)
        - if the string contains n opening and n closing brackets, 
the combination is complete and valid
            - add current string to the result list
            - return
        - else the current combination is incomplete
          - if left brackets count is less than n
            - increment left bracket count, add an opening bracket to the current string, and call dfs()
         - if right brackets count is less than left brackets count
           - increment right bracket count, add a closing bracket to the current string, and call dfs()
- start DFS with initial count of left and right brackets as 0 and an empty string
- return res


## Complexity
- Time complexity: O(recursion) → O(calls at each level ^ total levels) → O($2^n$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(recursion stack + result list) → O(recursion depth/levels + total valid combinations * each combination size) → O(n + Catalan number * n) → O(n + $C_n$ * n) → O($C_n$ * n)
    - recursion depth = total operations = n left + n right = 2n = n
    - total valid combinations = nth Catalan number because: <br>
    *n = 1 → 1 combination i.e. () <br>
    n = 2 → 2 combinations i.e. ()(), (()) <br>
    n = 3 → 5 combinations i.e. ((())), (()()), (())(), ()(()), ()()() <br>
    n = 4 → 14 combinations <br>
    1, 2, 5, 14, ... forms the Catalan sequence*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def generateParenthesis(self, n: int) -> List[str]:
        res = []
        
        def dfs(left, right, s):
            if left == right == n:
                res.append(s)
                return
            if left < n:
                dfs(left + 1, right, s + '(')
            if right < left:
                dfs(left, right + 1, s + ')')

        dfs(0, 0, '')
        return res
```
