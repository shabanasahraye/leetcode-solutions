## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. Simple way to solve 4Sum problem is to reduce it to 3Sum problem which can further be reduced to 2Sum problem. Just add an outer for-loop in the [3Sum solution](https://leetcode.com/problems/3sum/solutions/3527300/o-n-time-o-n-space-solution-explained).

2. Design an n-sum solution where n >= 2


## Approach
<!-- Describe your approach to solving the problem. -->
- sort the input array to handle duplicates easily
- set res = a list to store the result i.e unique quadruplets
- define a recursive helper function for nSum <br>
*nSum(start index, remaining numbers needed, remaining target, current chosen numbers)*
  - if only 2 numbers are needed
    - solve using [Two Sum II solution](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/solutions/3525831/on-time-o1-space-solution-explained-by-s-xc10/)
    - return after finishing 2Sum
  - else, more than 2 numbers are still needed
    - loop through the remaining numbers in the array <br>
    *start = starting index because previous numbers are already used <br>
     end = len(nums) - n + 1 to ensure enough numbers remain to complete the combination*
      - if current number is duplicate
        - skip it to avoid duplicate combinations
      - else choose current number and recursively solve (n-1)Sum
- call helper function to solve 4Sum
- return res


## Complexity
- Time complexity: O(sort + n_sum) → O(nlogn + n$^{n-1}$) → O(nlogn + n$^{4-1}$) → (nlogn + n$^3$) → O(n$^3$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        res = []

        def nSum(start, n, target, path):
            # base case: 2Sum 
            if n == 2:
                left, right = start, len(nums) - 1
                while left < right:
                    total = nums[left] + nums[right]
                    if total < target:
                        left += 1
                    elif total > target:
                        right -= 1
                    else:
                        res.append(path + [nums[left], nums[right]])
                        left += 1
                        right -= 1
                        # skip duplicates
                        while left < right and nums[left] == nums[left - 1]:
                            left += 1
                        while left < right and nums[right] == nums[right + 1]:
                            right -= 1
                return

            # reduce nSum to (n-1)Sum
            for i in range(start, len(nums) - n + 1):
                # skip duplicates
                if i > start and nums[i] == nums[i - 1]:
                    continue
                nSum(i + 1, n - 1, target - nums[i], path + [nums[i]])

        nSum(0, 4, target, [])
        return res
        
```
