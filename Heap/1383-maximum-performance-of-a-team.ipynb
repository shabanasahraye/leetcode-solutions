# <center> 1383. Maximum Performance of a Team </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-performance-of-a-team/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need a team of k engineers with max performance <br>
performance = sum(speed) * min(efficiency)

For minimum efficiency, sort the efficiency in descending order so that when we iterate we always get the least efficiency i.e efficiency of the current engineer will be less than the previous.

For speed sum, we need k engineers with max speed. We need to remove the engineers with the lowest speed. Use min heap because min heap pops the minimum element in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max performance
- set speed_sum = 0 to track the sum of k highest speeds
- create a min heap 
- pair corresponding efficiency and speed and sort the pairs in descending efficiency order
- traverse the sorted array
    - add current speed to heap and speed sum
    - if there are more than k elements in the heap
        - pop the lowest speed and subtract it from the speed sum
    - update performance by taking max of current performance and previous max performance
- mod res and return it


## Complexity
- Time complexity: O(sort + loop * heap push/pop) → O(nlogn + n * log(heap size)) → O(nlogn + n * logk) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort + heap) → O(n + k) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxPerformance(self, n: int, speed: List[int], efficiency: List[int], k: int) -> int:
        res = 0
        speed_sum = 0
        min_heap = []
        for e, s in sorted(zip(efficiency, speed), reverse=True):
            heappush(min_heap, s)
            speed_sum += s
            if len(min_heap) > k:
                speed_sum -= heapq.heappop(min_heap)
            res = max(res, speed_sum * e)
        return res % (10**9 + 7)
```
