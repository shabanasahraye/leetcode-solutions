## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
At every location, passengers either get in (start) or get out (end).
So instead of tracking each trip separately, track how many passengers are in the car at each point in time.

Approaches:
1. **Array** <br>
use a fixed-size array to simulate changes at each station. <br>
The time and space complexity will be constant, O(1000) → O(1), no matter the number of trips or stations.
2. **Heap** <br>
use a min heap to track active trips and remove finished ones. 
Sort the trips based on starting time to process the earlier trips first. Find the trips that end before a current trip to remove (drop off) those trips' passengers. To find the previous trip, use min heap because the previous trip end time is less than the current, and min heap returns min element in O(1) time.


# Approach
<!-- Describe your approach to solving the problem. -->
**Array Approach**
- create a list of size 1001 to track the number of onboard passengers at each station
    - (*total stations <= 1000, 1001 to keep the index inbound*)
- set total = 0 to track current passengers in the car
- traverse trips  
*for each trip passengers, start and end time*     
    - pick up passengers i.e add the current trip passengers to the start station
    - drop off passengers i.e remove the passengers from the end station
- traverse all stations i.e loop 1000 times
    - add the current station passengers to the total passengers count in the car
    - if total > capacity
      - return false
- if the loop completes without returning false, we can process all trips, return true

**Heap Approach**
- initialize a min heap to store active trips
  - *each entry = (trip ending time, passengers)*
- sort trips based on the start time
- set total = 0 to track current passengers in the car
- traverse trips <br>
*for each trip passengers, start and end time*
    - drop off the previous trip's passengers 
        - loop until heap top (previous trip end time) <= current trip start time
            - pop heap to get the ending time and passengers in the previous trip
            - subtract previous trip passengers from total
    - add current trip passengers to total
    - if total > car capacity, we can't process the current trip
      - return false
    - else we can process the current trip, push the current trip ending time and passengers to heap
- if the loop completes without returning false, we can process all trips, return true


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity:
    - Array Approach: O(creating at_station list + trips traversal + passengers traversal) → O(1000 + n + 1000) → n <= 1000 → O(1000 + 1000 + 1000) → O(1)
    - Heap Approach: O(sort + trip traversal * heap push pop) → O(nlogn + n * log(heap size)) → O(nlogn + n * logn) → O(nlogn)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity:
    - Array Approach: O(at_station list) → O(1000) → O(1)
    - Heap Approach: O(sort + heap) → O(n + n) → O(n)


## Code

```python
class Solution:

    def carPooling(self, trips: List[List[int]], capacity: int) -> bool:
        
        # Array Approach
        on_board = [0] * 1001
        total = 0
        for passengers, start, end in trips:
            on_board[start] += passengers
            on_board[end] -= passengers
        for passengers in on_board:
            total += passengers
            if total > capacity:
                return False
        return True


        # Heap Approach 
        # min_heap = []
        # trips.sort(key=lambda x: x[1])
        # total = 0
        # for passengers, start, end in trips:
        #     while min_heap and min_heap[0][0] <= start:
        #         end_time, p = heapq.heappop(min_heap)
        #         total -= p
        #     total += passengers
        #     if total > capacity:
        #         return False
        #     heapq.heappush(min_heap, (end, passengers))
        # return True
```
