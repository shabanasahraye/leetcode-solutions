## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To find Kth largest integer, we can use sorting or heap. <br>
We can't use them directly because it requires the numbers in integer form. The input is in string form and the string length can be up to 100. So, converting string to int is expensive. 

Once we define the correct ordering, we can use sorting directly on strings.

We can compare the numeric strings directly without conversion:
  - if two strings have different lengths, the longer string represents the larger number <br>
    for example '100' is greater than '99' <br>
  - if two strings have the same length,, lexicographical string comparison gives the correct order <br>
    for example '10' is smaller than '99' 


Approaches:
 - **Sorting** O(nlogn)<br>
  Define a custom comparison function using the above logic to sort the array. After sorting, the kth largest element is at index n - k.
  - **Heap** O(nlogk) <br>
   Use a min heap of size k to store the k largest numbers seen so far. If the heap size exceeds k, remove the smallest number. After processing all numbers, the heap top is the kth largest number.


## Approach
<!-- Describe your approach to solving the problem. -->
**Sorting Approach**
- *there are some edge cases but handling them here is expensive because the numbers are in string form, they are auto handled by the sorting and heap*
    - *if k = 1, kth largest element is max(Array)*
    - *if k = size of the array, the kth largest element is min(array)*
- define a custom comparison function
    - a comparison function returns:
        - -1 if the first value is smaller
        - +1 if the first value is greater
        - 0 if both are equal
    - custom function cmp(num1, num2)
        - if num1 and num2 length is different
            - return -1 if num1 length is less else +1
        - else the lengths are the same, compare the strings
          - return a negative value if num1 < num2
          - return a positive value if num1 > num2
          - return 0 if num1 == num2
- convert the comparison function to a key function using cmp_to_key() and sort the number using the key function
- return the number at index n-k
  - *(after sorting, kth largest number will be at index n - k)*

**Heap Approach**
- initialize a min heap
- traverse all numbers
  - push (len(num), num) into the heap
    - *(tuple because tuple comparison automatically compares length first and then the string)*
  - if heap size becomes greater than k
    - pop the heap top
      - *(this removes the smallest number among the current k + 1 numbers)*
- return the string part of the heap top
  - *(after processing all numbers, the heap contains the k largest numbers, the heap top is the kth largest number)*



## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: 
  - Sorting Approach: O(sort) → O(nlogn)
  - Heap Approach: O(array traversal * heap push/pop) → O(n * logk) → O(nlogk)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: 
  - Sorting Approach: O(sort) → O(n)
  - Heap Approach: O(heap size) → O(k)
