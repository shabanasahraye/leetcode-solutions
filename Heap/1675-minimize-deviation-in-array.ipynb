# <center> 1675. Minimize Deviation in Array </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to minimize the max deviation (i.e max element - min element).

To minimize deviation:
  - we can multiply odd numbers only once (to make them even)
  - divide even numbers multiple times (to reduce them)

To minimize the deviation, we can perform the multiplication operation on odd numbers only once and the division operation on even numbers multiple times.

To normalize the array:
  - first convert all numbers to their maximum possible value i.e multiply odd numbers by 2
  - then repeatedly reduce the maximum element if it is even i.e divide by 2

We need to track the:
  - max element because it will change due to division, use max heap because it returns the max element in O(1) time
  - current minimum element since deviation depends on both ends


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = max possible number to track the minimum deviation
- set mini = max possible number to track the current minimum element in the array
- create a max heap 
    - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- traverse the array
    - if the number is odd, multiply it by 2
    - flip the number sign and push it to heap
    - update mini
- loop while the deviation is not 0 or the max element is even
    - pop heap to get the max element and flip the sign because we have added the inverted values to the heap
    - update deviation by taking minimum of current deviation and previous min deviation
    - if deviation is 0 or max element is odd
      - break the loop because we cannot reduce further
    - divide max element by 2
    - update the mini
    - push the updated max element back into heap
- return res


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(for loop * heap push + while loop * heap push/pop) → O(n * log(heap size) + n * log(heap size)) → O(n * logn + n * logn) → O(n * logn)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(heap size) → O(n)


## Code

```python
class Solution:

    def minimumDeviation(self, nums: List[int]) -> int:
        res = math.inf
        mini = math.inf
        max_heap = []
        for num in nums:
            if num & 1:
                num *= 2
            heappush(max_heap, -num)
            mini = min(mini, num)
        while True:
            maxi = -heapq.heappop(max_heap)
            res = min(res, maxi - mini)
            if res == 0 or maxi & 1:
                break
            maxi //= 2
            mini = min(mini, maxi)
            heappush(max_heap, -maxi)
        return res
```
