# <center> 1882. Process Tasks Using Servers </center>


## Problem Description
[Click here](https://leetcode.com/problems/process-tasks-using-servers/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given m number of tasks with processing time and n number of servers with weights. We have to schedule these tasks and return which task will be assigned to which server i.e the index of the server assigned to each task.

For each task arriving at time t:
- release all busy servers with availability time <= t
- if free servers are available, pick the one with minimum weight and index and assign it to the task
- if no free server is available, release the earliest available busy server and assign it to the task 

We need to track the minimum weight in free servers and the earliest available time in busy servers. Use min heap because it returns the min element in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the result i.e the indices of the assigned servers
- initialize a min heap for busy servers
    - *for each server the heap will contain availability time, weight, and index i.e the heap is sorted by availability time*
- create a min heap for free servers
    - for each server, the heap will contain weight, index, and availability time i.e the heap is sorted by weight and then index
    - *create a list first and then heapify to create a min heap. If you create min-heap without heapify i.e by pushing elements one by one, the time complexity will be O(nlogn)*
- traverse all tasks <br>
*for each task index (start time) and value (processing time)*
    - if there are busy servers and busy server availability time <= start time of the current task
        - release the busy server by removing it from the busy server heap and adding it to the free server heap 
    - if there is a server available
        - pop free server heap to get the minimum weight server
        -  assign it to process the task by marking it busy i.e push it to the busy heap 
        - add its index to the result list
    - else there is no server available
        - pop busy server heap to get the busy server with the min (earliest) availability time
        - assign it to the task i.e push the server with its new availability time to the busy heap 
        - add its index to the result list
- return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(free server list + heapify + tasks traversal * heap push/pop) → O(n + n + m * log(heap size)) → O(n + n + m * log(n)) → O(mlogn)
    - *n = number of servers*
    - *m = number of tasks*
    - *heap size = total servers = n*
    - *the inner while loop will not run n times for each outer loop iteration because each server is moved from busy heap to free heap only when it becomes available*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(free server heap + busy server heap + result list) → O(n + n + m) → O(n + m)
    - *free server heap and busy server heap together store at most n servers*
    - *result list stores one server index for each task, therefore its size is m*


## Code

```python
class Solution:

    def assignTasks(self, servers: List[int], tasks: List[int]) -> List[int]:
        res = []
        busy = []
        free = [(weight, i, 0) for i, weight in enumerate(servers)]
        heapify(free)
        for start_time, process_time in enumerate(tasks):
            while busy and busy[0][0] <= start_time:
                free_time, weight, index = heappop(busy)
                heappush(free, (weight, index, free_time))
            if free:
                weight, index, free_time = heappop(free)
                res.append(index)
                heappush(busy, (start_time + process_time, weight, index))
            else:
                free_time, weight, index = heappop(busy)
                res.append(index)
                heappush(busy, (free_time + process_time, weight, index))
        return res
```
