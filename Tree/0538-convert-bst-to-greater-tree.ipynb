# <center> 538. Convert BST to Greater Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/convert-bst-to-greater-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each node value in a greater tree = node value + sum of all nodes greater than it i.e the postfix sum.

Use DFS with reverse inorder traversal (right → node → left) to visit nodes in descending order to compute the postfix sum.
Keep a running sum, and update each node by adding its value to this sum.

Reverse inorder traversal processes all greater nodes before the current node, so the running sum always contains all required greater values.


## Approach
<!-- Describe your approach to solving the problem. -->
- set post_sum = 0 to track the postfix sum
- use a helper function for DFS <br>
*dfs(current node)*
    - make post_sum variable nonlocal 
        - *(this is to access the outer function variable in nested function)* 
    - if the current node is not null
        - traverse the right subtree first to find greater nodes
          - *(we are doing reverse inorder traversal: right → node → left so that we always visit greater nodes first)*
        - add current node value to postfix sum
        - replace the current node value with the postfix sum
        - traverse the left subtree
- start dfs() from the root node
- return root


## Complexity
- Time complexity: O(DFS) → O(total nodes visited) → O(n)
  - *each node is visited exactly once during DFS*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
  - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def convertBST(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        post_sum = 0

        def dfs(node: Optional[TreeNode]) -> None:
            nonlocal post_sum
            if node:
                dfs(node.right)
                post_sum += node.val
                node.val = post_sum
                dfs(node.left)

        dfs(root)
        return root
```
