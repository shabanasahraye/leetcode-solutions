# <center> 116. Populating Next Right Pointers in Each Node </center>


## Problem Description
[Click here](https://leetcode.com/problems/populating-next-right-pointers-in-each-node/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to set the next pointer of each node to its next node on the same level. Use the next pointers to traverse the current level and connect the nodes in the next level in O(1) space instead of using BFS. 

Move level by level from left to right on each level and set the next pointer of each node to its next node.

The tree is a perfect binary tree i.e each node has two children.
- For the left node there's always a next node i.e the right node
- For the right node, there may or may not be a next node. The right node has a next node if its parent node has a sibling.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base case
    - if there's no node, return null
- set cur = root node to track the current node
- set nxt = root left child to track the leftmost node of the next level
  - *(if nxt becomes null, current level is the last level)*
- loop until the last level
    - set the current node left child next to the current node right child 
    - if the current node has a sibling, the right child has a next node
        - set right child next to the sibling node left child
        - update cur i.e set it to its sibling node
    - else the right child has no next node, it is the last node of the current level
        - the last node's next pointer should be null but we don't need to set it to null because the next pointer's default value is null
        - update cur and nxt to move to the next level
          - *(cur will point to the leftmost node of the next level and nxt will point to the leftmost node of the level below it)*
- return root


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(tree traversal) → O(total nodes visited) → O(n)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(1)


## Code

```python
"""
# Definition for a Node.
class Node:

    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""


class Solution:

    def connect(self, root: 'Optional[Node]') -> 'Optional[Node]':
        if not root:
            return None
        cur, nxt = root, root.left 
        while cur and nxt:
            cur.left.next = cur.right
            if cur.next:
                cur.right.next = cur.next.left
                cur = cur.next
            else:
                cur = nxt
                nxt = cur.left
        return root
```
