## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
inorder traversal = left → root → right

Use stack to simulate inorder traversal of a BST.

The stack always stores the path of left nodes from the current subtree root, so that the next smallest element is always on top.


## Approach
<!-- Describe your approach to solving the problem. -->

**init()**
- create a stack and add all left nodes to stack
  - *(this stores the path to the smallest element)*

**next()**
- pop the top node (i.e next smallest element) from the stack
- if the popped node has a right child
  - add the right child and all its left descendants to the stack
- return the popped node value

**hasNext()**
- check if stack is empty and return the result


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity:
    - init(): O(h) → O(n)
        - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
    - next(): amortized O(1)
        - *each node is pushed and popped at most once across all calls, so total work is O(n)*
        - a single call may take O(h), but this cost is distributed over all nodes, making the average O(1)
    - hasNext(): O(1)

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: 
    - init(): O(h) → O(n)
    - next(): O(h) → O(n)
    - hasNext(): O(1)


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class BSTIterator:

    def __init__(self, root: Optional[TreeNode]):
        self.stack = []
        while root:
            self.stack.append(root)
            root = root.left

    def next(self) -> int:
        res = self.stack.pop()
        root = res.right
        while root:
            self.stack.append(root)
            root = root.left
        return res.val

    def hasNext(self) -> bool:
        return self.stack
        


# Your BSTIterator object will be instantiated and called as such:
# obj = BSTIterator(root)
# param_1 = obj.next()
# param_2 = obj.hasNext()
```
