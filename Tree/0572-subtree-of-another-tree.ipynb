## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given two trees A *(root)* and B *(subRoot)*. We need to check if a subtree of A is the same as B.

Traverse the tree A. For every node in tree A, check if the subtree rooted with this node is the same as tree B.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if tree B is empty, return true because a tree always has an empty subtree. No need to check this condition because according to the constraints, tree B is never null
  - if tree A is empty or becomes empty (i.e we have reached the end of a subtree during recursion), return false because an empty tree can't have a non-empty subtree
- if the current node tree (subtree of A) is the same as tree B, return true
    - *(to check whether trees are the same or not, define another function using [Same Tree Solution](https://leetcode.com/problems/same-tree/solutions/3628991/o-min-m-n-time-o-min-m-n-space-solution-explained/))*
- else check the left and right subtrees and return true if there is a subtree same as tree B either in the left or right


## Complexity
- Time complexity: O(tree A traversal * comparing subtree of tree A with tree B) → (tree A size * tree B size) → O(n * m)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
  - *recursion depth is bounded by the height of the trees*
  - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        # if not subRoot:
        #     return True
        if not root:
            return False
        if self.is_same_tree(root, subRoot):
            return True
        return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)

    def is_same_tree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q:
            return True
        if not p or not q or p.val != q.val:
            return False
        return self.is_same_tree(p.left, q.left) and self.is_same_tree(p.right, q.right)
```
