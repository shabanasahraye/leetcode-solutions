# <center> 450. Delete Node in a BST </center>


## Problem Description
[Click here](https://leetcode.com/problems/delete-node-in-a-bst/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Search for the node in the tree. If the node to be deleted is a:
- leaf node → remove it
- node with a single child → replace it with the child node
- node with 2 children → replace it with the minimum value node in the right subtree or the maximum value node in the left tree 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the root is null, we have reached the tree end, return null
- if the node to be deleted is less than the current node
  - search for the node in the left subtree and delete it 
- if the node to be deleted is greater
  - search for the node in the right subtree and delete it
- else current node is the node to be deleted
    - if the current node has no left child
      - return its right child
    - if the current node has no right child
      - return its left child
    - else the node has two children
        - find the min value node in the right subtree
        - replace the current node value with the min value
        - now there will be two same nodes but duplicates are not allowed in BST, delete the duplicate node (the min value node) from the right subtree
- return the root


## Complexity
- Time complexity: O(tree traversal) → O(h) → O(n)
  - *O(h) because this is a BST, so at each node we move to either the left or right subtree and only traverse one path downward*
  - *height (h) depends on tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(logn)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        if not root:
            return None
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            if not root.left:
                return root.right
            elif not root.right:
                return root.left
            cur = root.right
            while cur.left:
                cur = cur.left
            root.val = cur.val
            root.right = self.deleteNode(root.right, root.val)
        return root
```
