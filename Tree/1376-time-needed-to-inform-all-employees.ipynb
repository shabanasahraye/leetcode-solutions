## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is a tree problem. Consider the head of the company as the root of a tree and direct subordinates as child nodes. 

The time required to inform all employees is the time to reach from root node to distant leaf node i.e we have to find the longest path from root to leaf node. To find longest path, use DFS.

Build the tree from the manager array, traverse each node (employee) and compute the time to reach the root (head).

Note: The value in manager array is manager id and the index is employee id.


## Approach
<!-- Describe your approach to solving the problem. -->
- build the hierarchy tree using an adjacency list
  - create a default dict to represent tree
    - *key = manager (node)*
    - *value = list of employees indices (sub-nodes)*
  - for each employee `i`
    - if it is not the root (company head) i.e it has a manager
      - add the employee to the employee list of its manager
        - *(this builds a parent → children relationship in the tree)*
- define a helper function for DFS <br>
  *dfs(current node)*
  - set time = 0 to track the max time needed to inform all sub-nodes from the current node
  - recursively compute the time needed by all its children and track the maximum time
  - return total time for this subtree rooted at current node
    - *total time = max time from children + current node informTime*
- start DFS from the root i.e headID and return the result
  - *this gives the total time required to inform all employees in the company*



## Complexity
- Time complexity: (buildign tree + tree DFS) → O(adjacency list + total nodes visited) → O(n + n) → O(n)
  - *each node is visited exactly once during DFS*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(tree + recursion stack) → O(adjacency list + h) → O(n + n) → O(n)
  - *one node is created for each employee*
  - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    def numOfMinutes(self, n: int, headID: int, manager: List[int], informTime: List[int]) -> int:
        tree = defaultdict(list)

        # build tree (manager → employees)
        for i in range(n):
            if manager[i] != -1:
                tree[manager[i]].append(i)

        def dfs(node: int) -> int:
            time = 0
            for child in tree[node]:
                time = max(time, dfs(child))
            return time + informTime[node]

        return dfs(headID)
```
