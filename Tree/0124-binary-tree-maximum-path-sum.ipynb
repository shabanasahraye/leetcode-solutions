# <center> 124. Binary Tree Maximum Path Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/binary-tree-maximum-path-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The max path sum (i.e max gain) passing through a node is the max of:
1. node value 
2. node value + left subtree max gain
3. node value + right subtree max gain
4. node value + left subtree max gain + right subtree max gain

Use DFS to find the maximum gain from the left and right subtrees.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = root node value or negative infinity to track the maximum path sum 
- define a helper function for DFS <br>
*dfs(current node)*
    - if the node is null, return 0
    - make res nonlocal
    *(this is to access the outer function variable in nested function)*
    - get the max path sum in the left subtree
    - get the max path sum in the right subtree
    - update res with the max path sum passing through the current node
      - *node value + left subtree max path sum + right subtree max path sum*
    - return the maximum path sum that can be extended to the parent node
      - *max(sum of current node and max of left and right subtree path sum, 0)*
      - *0 because a node value can be negative*
      - *the DFS return value ignores negative paths by returning 0, so this automatically handles all cases: 
      node value <br>
      node value + left path sum <br>
      node value + right path sum <br>
      node value + left path sum + right path sum*
- start the dfs
- return res


## Complexity
- Time complexity: O(tree DFS) → O(total nodes visited) → O(n)
  - *each node is visited exactly once during DFS*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        res = root.val
        
        def dfs(node: Optional[TreeNode]) -> int:
            if not node:
                return 0
            nonlocal res
            left_sum = dfs(node.left)
            right_sum = dfs(node.right)
            res = max(res, node.val + left_sum + right_sum)
            return max(node.val + max(left_sum, right_sum), 0)
        
        dfs(root)
        return res
```
