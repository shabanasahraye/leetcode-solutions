### Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A binary search tree (BST) is valid if:
- all nodes in the left subtree are smaller than the root
- all nodes in the right subtree are greater than the root
- both subtrees are also valid BSTs
- there is no duplicate node

Do DFS to validate each node using range constraints (left bound and right bound), instead of explicitly finding min/max in subtrees.

BST is valid if every node lies in a valid (min, max) range or inorder traversal is strictly increasing.


## Approach
<!-- Describe your approach to solving the problem. -->
**recursive approach**
- define a helper function for DFS <br>
dfs(current node, left bound, right bound)
    - if current node is null
      - return true because we have reached the tree end and the traversal is completed without returning false i.e BST is valid
    - return false if current node is not valid i.e current node val is not within valid range
    - else the current node is valid
        - recursively validate the left and right subtrees with updated bounds and return the result
- start dfs from root with left bound as min possible value and right bound as max possible value

**iterative approach**
- create a stack for DFS (inorder traversal)
- set prev = null to track the previously visited node in the left subtree
- set cur = root node to track the current node
- loop while the stack is not empty or current node is not null
    - loop till the end of left subtree
        - push current node to the stack
        - move to the left child
    - pop the stack top and update current node 
    - if current node value is less than prev node i.e it is not in increasing order, the current node is invalid
        - return false because the tree is not a BST
    - else the current node is valid
        - update prev i.e set current node as previous node
        - update cur i.e move to the right subtree
- if the loop completes without returning false, BST is valid, return true


## Complexity
- Time complexity:
    - Recursive Approach: O(n)
      - *each node is visited exactly once*
    - Iterative Approach: O(n)
      - *each node is pushed onto the stack once and popped once*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity:
    - Recursive Approach: O(recursion stack) → O(h) → O(n)
        - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative Approach: O(stack) → O(h) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->
