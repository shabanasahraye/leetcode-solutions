# <center> 951. Flip Equivalent Binary Trees </center>


## Problem Description
[Click here](https://leetcode.com/problems/flip-equivalent-binary-trees/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Two trees, A and B, are flip equivalent if:
- root nodes have the same value and
- left subtree of A is flip equivalent to the left subtree of B and right subtree of A is flip equivalent to the right subtree of B or
- left subtree of A is flip equivalent to the right subtree of B and right subtree of A is flip equivalent to the left subtree of B

Do DFS to traverse both trees simultaneously and check the node values and subtrees.


## Approach
<!-- Describe your approach to solving the problem. -->
- if both tree nodes are null
  - return true because they are flip equivalent
- if only one node is null or the node values differ
  - return false because they are not flip equivalent
- else both nodes are not null and have the same value
  - recursively check their subtrees for:
    - no flip: left with left and right with right
    - flip: left with right and right with left
  - return the result i.e true if either case is flip equivalent else false


## Complexity
- Time complexity: O(tree DFS) → O(total nodes visited) → O(n)
  - each node is visited and compared at most once during DFS
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(max height) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def flipEquiv(self, root1: Optional[TreeNode], root2: Optional[TreeNode]) -> bool:
        if not root1 and not root2:
            return True
        if not root1 or not root2 or root1.val != root2.val:
            return False
        return self.flipEquiv(root1.left, root2.left) and self.flipEquiv(root1.right, root2.right) or self.flipEquiv(root1.left, root2.right) and self.flipEquiv(root1.right, root2.left)
```
