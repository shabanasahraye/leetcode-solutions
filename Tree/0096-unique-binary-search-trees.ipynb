## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use the Catalan number formula to find the number of unique binary search trees (BSTs) with n nodes because the number of unique BSTs follows the Catalan sequence.

For example: <br>
0 node → 1 BST <br>
1 node → 1 BST <br>
2 nodes → 2 BST <br>
3 nodes → 5 BST <br>
4 nodes → 14 BST <br>
5 nodes → 42 BST

1, 1, 2, 5, 14, 42... → Catalan sequence

Formula to find the nth Catalan number:

${\displaystyle C_{n+1}=\sum _{i=0}^{n}C_{i}\,C_{n-i}\quad {\text{for }}n\geq 0}$ <br>
${\displaystyle C_{n}=\sum _{i=1}^{n}C_{i-1}\,C_{n-i}}$

count(n) = Σ(count(left) × count(right)) <br>
count(n) = Σ(count(i - 1) × count(n - i)), where i is chosen as the root
- i is the root
- nodes in left subtree = values 1 to i - 1 → total nodes = i - 1
- nodes in right subtree = values i + 1 to n → total nodes = n - i
- i range = 1 to n


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize an array of size n + 1 with zeroes
  - *dp[i] stores the number of unique BSTs that can be formed using i nodes*
- set the elements at index 0 and 1 to 1 (see Catalan sequence)
  - *dp[0] = 1 because an empty tree is a valid BST*
  - *dp[1] = 1 because a single node can form only one BST*
- loop in range 2 to n <br>
*i represents the total number of nodes used to form BSTs*
  - try every possible root position j from 1 to i
    - multiply the unique BSTs counts from left and right subtrees, and add it to the total count at index i
- return value at index n


## Complexity
- Time complexity: O(outer loop * inner loop) → O(n²)
    - *n² because for every outer iteration i, the inner loop runs i times*
    - *for example, for outer loop iteration 2 the inner loop runs 2 times* <br>
    *for outer loop iteration 3 the inner loop runs 3 times, and so on* <br>
    *total iterations = 2 + 3 + 4 + ... + n = n(n + 1)/2 - 1 = (n² + n)/2 ~ n²*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: (dp array) → O(n)
  - *the dp array stores the number of unique BSTs for every value from 0 to n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numTrees(self, n: int) -> int:
        dp = [0] * (n + 1)
        dp[0] = dp[1] = 1
        for i in range(2, n + 1):
            for j in range(1, i + 1):
                dp[i] += dp[j - 1] * dp[i - j]     # or dp[i] += dp[j] * dp[i - j - 1] 
        return dp[n]
```
