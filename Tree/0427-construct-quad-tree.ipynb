# <center> 427. Construct Quad Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/construct-quad-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a quad tree, each node has either 0 or 4 children. The quad tree represents a grid (2D array).

Recursively divide the grid into 4 quadrants (top-left, top-right, bottom-left, bottom-right) and construct a quad tree for each quadrant. If all values in a quadrant are the same, create a leaf node. Else create a non-leaf node and set its children as the quad trees for each quadrant.


## Approach
<!-- Describe your approach to solving the problem. -->
- define a recursive DFS function dfs(n, r, c) where:
  - *n = size of current sub-grid*
  - *r, c = top-left corner of current sub-grid*
  - *for each sub-grid:*
    - set flag to true to track if all values in the current n x n region are the same
    - loop through the grid region to check if all values are the same and toggle flag when mismatch is found
    - if flag is true i.e all values are the same
        - create a leaf node with that value and return <br>
        *(this means the entire region is uniform, so no need to split further)*
    - else values are not the same
      - split into 4 equal quadrants (top-left, top-right, bottom-left, bottom-right) and recursively build a quad tree for each quadrant
      - return a non-leaf node with these 4 children
- start DFS from the full grid (n = len(grid), r = 0, c = 0)


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(recursion depth/levels * work per level) → O(logn * n²) → O(n²logn)
  - *recursion depth is determined by how many times the grid can be divided until we reach a 1 × 1 cell <br>
    each step reduces the grid size by half i.e from n → n/2 → n/4 → ... → 1 <br>
    so depth = O(log n)*
  - *n² to check if all elements in the current region are the same*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(recursion stack) → O(recursion depth) → O(logn)


## Code

```python
"""
# Definition for a QuadTree node.
class Node:
    def __init__(self, val, isLeaf, topLeft, topRight, bottomLeft, bottomRight):
        self.val = val
        self.isLeaf = isLeaf
        self.topLeft = topLeft
        self.topRight = topRight
        self.bottomLeft = bottomLeft
        self.bottomRight = bottomRight
"""


class Solution:

    def construct(self, grid: List[List[int]]) -> 'Node':
        
        def dfs(n, r, c):
            same = True
            for i in range(n):
                for j in range(n):
                    if grid[r][c] != grid[r + i][c + j]:
                        same = False
                        break
            if same:
                return Node(grid[r][c], True)
            n = n // 2
            top_left = dfs(n, r, c)
            top_right = dfs(n , r , c + n)
            bottom_left = dfs(n , r + n , c)
            bottom_right = dfs(n , r + n , c + n)
            return Node(0, False, top_left, top_right, bottom_left, bottom_right)

        return dfs(len(grid), 0, 0)
```
