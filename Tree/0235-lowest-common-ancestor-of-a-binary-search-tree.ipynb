## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find the lowest common ancestor (LCA) of nodes p and q.

Traverse the binary search tree (BST). If a node value is:
- greater than both p and q, then the LCA lies in the left subtree
- smaller than both p and q, then the LCA lies in the right subtree
- else the current node is the LCA


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- if current node value is greater than p and q, search the left subtree for LCA
- if current node value is smaller than p and q, search the right subtree for LCA
- else the current node is the LCA, return it

**Iterative Approach**
- set cur = root to track the current node
- loop until the end of the tree
    - if current node value is greater than p and q, set cur to the left child to search in the left subtree
    - if the current node value is smaller, set cur to the right child to search in the right subtree
    - else the current node is the LCA, return it


## Complexity
- Time complexity: 
  - Recursive Approach: O(tree traversal) → O(h) → O(n)
    - *O(h) because this is a BST, so at each node we move to either the left or right subtree and only traverse one path downward*
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach: O(tree traversal) → O(h) → O(n)

- Space complexity:

  - Iterative Approach: O(1)
