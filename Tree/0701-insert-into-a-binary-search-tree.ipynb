## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a binary search tree (BST), left node values are smaller, and right node values are greater than the root node.

To insert a new value (node) in a binary search tree, find a leaf node that satisfies the insertion condition and insert the new node as a child of the leaf node.
  - left child if new node value < leaf node value
  - right child if new node value > leaf node value


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
    - if current node is null, either the tree is empty or we have reached the insertion position
      - create a new node and return it 
- if the new value is less than the current node value
  - search for the leaf node in the left subtree and insert
- else value is greater
  - search the right subtree and insert
- return the root node

**Iterative Approach**
- check base cases
    - if current node is null, either the tree is empty or we have reached the insertion position
      - create a new node and return it 
- set current node pointer cur = root
- set leaf node pointer leaf = null
- loop until we find the leaf node for insertion
    - update the leaf to be the current node
    - if the value is less than the current node
      - go to the left subtree
    - else go to the right subtree
- if the value is less than the leaf node
  - create a left child node
- else create a right child node
- return root


## Complexity
- Time complexity: 
    - Recursive Approach: O(tree traversal) → O(h) → O(n)
      - *O(h) because this is a BST, so at each node we move to either the left or right subtree and only traverse one path downward*
      - *height (h) depends on tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative Approach: O(tree traversal) → O(h) → O(n)
