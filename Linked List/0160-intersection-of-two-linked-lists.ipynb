## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. O(n) space <br>
Traverse the first list and mark its nodes visited by adding them to a hashset. Traverse the second list, if a node is already present in the set, it is the intersection node.
2. O(1) space <br>
Use two pointers starting at each list head. When a pointer reaches the end, redirect it to the other list’s head. By traversing both lists, both pointers cover the same total distance, so they meet at the intersection if it exists.


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = head of list1 to traverse list1
- set j = head of list2 to traverse list2
- loop until both pointers point to the same node, this can be either the intersection node or null if no intersection exists
    - if we have not reached list1 end
        - move i to the next node in list1 
    - else set i to the head of list2
    - if we have not reached list2 end
        - move j to the next node in list2 
    - else set j to head of list1
- return i or j <br>
*(because both will point to the same node or null if no intersection exists)*


## Complexity
- Time complexity: O(loop) →  O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None


class Solution:

    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        i, j = headA, headB
        while i != j:
            i = i.next if i else headB
            j = j.next if j else headA
        return i
```
