# <center> 25. Reverse Nodes in k-Group </center>


## Problem Description
[Click here](https://leetcode.com/problems/reverse-nodes-in-k-group/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Find the total number of groups (sublists) and reverse each group.

To reverse a sublist, use the Pick-Cut-Front logic i.e each time take the next node (node after the current node), remove it from its position, and insert it at the front (i.e after previous node).

Example: 1 → 2 → 3
```
# initial
prev → 1 → 2 → 3
      cur → 1
      cur.next → 2

# pick 2 → cut it → move to front
prev → 2 → 1 → 3
      cur → 1
      cur.next → 3

# pick 3 → cut it → move to front
prev → 3 → 2 → 1
      cur → 1
      cur.next → null
```


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if k is 1, return because the result after reversing will be the same as the original list
- create a dummy node and point it to the head
- set current node pointer cur = head
- set previous node pointer prev = dummy
- set n = 0 to find the length of the list (i.e total nodes)
- loop until the last node
  - increment n
  - update cur to move to the next node
- reset cur to point to the head node
- find total number of groups i.e n/k and loop <br>
*for each group*
  - loop k - 1 times to reverse the k-node sublist using pick-cut-front logic
    - take the next node
    - remove it from its position
    - insert it at the front of the group i.e after prev by updating the next pointers 
  - update prev to cur (tail of the reversed group)
  - move cur to next node (start of next group)
- return dummy.next as the head of the modified list


## Complexity
- Time complexity: O(finding list size + loops to reverse groups)  → O(n + n/k * k) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if k == 1:
            return head
        dummy = ListNode(0, head)
        cur = head
        prev = dummy
        n = 0
        while cur:
            n += 1
            cur = cur.next
        cur = head
        for _ in range(n // k):
            for _ in range(k - 1):
                nextt = cur.next
                cur.next = nextt.next
                nextt.next = prev.next
                prev.next = nextt
            prev = cur
            cur = cur.next
        return dummy.next
````
