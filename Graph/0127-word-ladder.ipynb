# <center> 127. Word Ladder </center>


## Problem Description
[Click here](https://leetcode.com/problems/word-ladder/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the shortest path from the start word to the end word. 

To find the shortest path in a graph, use Breadth First Search (BFS).


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if start and end words are the same
        - return 0
        - *(no need to check this because no test cases have been added for it)*
    - if end word is not in the word list, there is no path
      - return 0
- create a hashmap to represent the graph (adjacency list)
    - *key = node*
    - *value = list of neighbor nodes*
- create a queue for BFS
- create a hashset to mark visited nodes
- add the start word to the word list
- create/fill the graph. The adjacent words in the path from the start word to the end word should differ by one character. We will find the intermediate words of all the words and treat them as graph nodes. For example, for the word 'BOOK', the intermediate patterns are '\*OOK', 'B\*OK', 'BO\*K', and 'BOO\*'
    - for each word
        - for each character index in the word
            - find the intermediate word
              - *intermediate word = prefix + * + postfix*
            - add an edge from the word to the intermediate pattern i.e add the word to the neighbors' list of the intermediate pattern
- add (start word, distance) to the queue
- mark the start word as visited
- Do BFS
    - loop until q is empty
        - pop q to get a node/word and distance to reach that node
        - if the node is the same as the end word, we get the shortest path, return the distance
        - else find the intermediate words for the node and add unvisited nodes to the queue with updated distance and mark them visited
- if BFS completes without returning anything, there is no path, return 0

## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(wordList check for endWord + graph (adjacency list) creation + BFS) → O(total words + generating all intermediate patterns + generating intermediate patterns during BFS) → O(total words + total words * total patterns * pattern creation cost + total words * total patterns * pattern creation cost) →  O(N + (N × L × L) + (N × L × L)) → O(N + N × $L^2$ + N × $L^2$) → O(N × $L^2$)
    - *N = number of words in wordList*
    - *L = length of each word*
    - *for each word, we generate L intermediate patterns, and each pattern creation takes O(L) time*
    - *the complexity is expressed in terms of N and L instead of the graph notation V and E because the graph is implicit (generated from word transformations) rather than explicitly given*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(graph(adjacency list) + BFS queue + seen hashset) → O(total intermediate patterns + total words in queue + total visited words) → O(total words * total patterns per word + total words + total words) → O(N × L + N + N) → O(N × L)


## Code

```python
class Solution:

    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        if endWord not in wordList:
            return 0
        graph = defaultdict(list)
        q, seen = deque(), set()
        wordList.append(beginWord)
        for word in wordList:
            for i in range(len(word)):
                pattern = word[:i] + '*' + word[i + 1:]
                graph[pattern].append(word)
        q.append((beginWord, 1))
        seen.add(beginWord)
        while q:
            word, dist = q.popleft()
            if word == endWord:
                return dist
            for i in range(len(word)):
                node = word[:i] + "*" + word[i + 1:]
                for neighbor in graph[node]:
                    if neighbor not in seen:
                        seen.add(neighbor)
                        q.append((neighbor, dist + 1))
        return 0  
```
