## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given an undirected connected graph represented as an adjacency list and need to create a deep copy of it.

Traverse the graph using DFS and copy each node. Since graphs can contain cycles, use a hashmap to track the visited/copied nodes.


## Approach
<!-- Describe your approach to solving the problem. -->
- if there are no nodes
  - return null
- create a hashmap
  - *key = original node*
  - *value = copied node*
- define a helper function for DFS <br>
*dfs(current node)*
    - if current node is already visited (copied)
      - return its copy node from the hashmap
    - create a copy the current node 
    - add the original node and its copy to the hashmap
    - for each neighbor of the current node
        - recursively copy the neighbor node and add it to the neighbors list of the current copy
    - return the current node copy
- start DFS from the given node and return its copy


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(graph traversal) → O(adjacency list DFS) → O(V + E)
    - *V = total vertices/nodes, represented by the rows/keys in the adjacency list*
    - *E = total edges, represented by the neighbor lists in the adjacency list*
    - *each node is visited once and each edge is processed once during DFS*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(hashmap + DFS recursion stack) → O(total nodes + recursion depth) → O(V + V) → O(V)
  - *hashmap stores one cloned node for each original node*
  - *recursion depth is V because DFS may visit every node before returning*


## Code

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""


class Solution:

    def cloneGraph(self, node: 'Node') -> 'Node':
        if not node:
            return None
        clone = {}

        def dfs(node: 'Node') -> 'Node':
            if node in clone:
                return clone[node]
            copy = Node(node.val)
            clone[node] = copy
            for n in node.neighbors:
                copy.neighbors.append(dfs(n))
            return copy

        return dfs(node)
```
