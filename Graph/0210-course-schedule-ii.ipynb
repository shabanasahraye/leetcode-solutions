## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Course Schedule I](https://leetcode.com/problems/course-schedule/), the difference is that we need to return the order of courses (topological order). 

Create a directed graph with courses as vertices/nodes and prerequisite relationships as edges. <br>
Do DFS for cycle detection and topological sort. If there's no cycle, we can finish all the courses.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap for the graph (adjacency list)
    - *key = node (course)*
    - *value = list of neighbors (prerequisite courses)*
- create a hashmap to track the status of a node
    - *key = node (course)*
    - *value = status (0 = unvisited, -1 = visiting i.e on the current path, 1 = visited)*
- create an empty list to store the courses in order
- fill the hashmap to create graph
    - for each edge (course, prerequisite) in the prerequisites list
        - add the prerequisite to the neighbors list of the course
            - *this creates an edge from the course to its prerequisite*
- define a helper function for DFS <br>
*dfs(current node/course)*
    - if status is -1, we encountered the current node again in the path, there is a cycle
        - return false
    - if status is 1, current node is visited and cycle-free (all of its prerequisite paths have been checked)
        - return true
    - else it is an unvisited node
        - add it to the current path by updating the status to -1
        - visit all the neighbors to check for cycle
            - if there is a cycle and we can't finish the course i.e dfs returns false
                - return false
        - if the neighbors loop executes without returning false, the current node is cycle-free and we can finish this course
            - remove it from the current path by updating the status to 1
            - add it to the result list
                - *the course is added after its prerequisites, producing a valid topological order*
            - return true 
- for each course
    - do DFS to check if we can finish this course. if we can't i.e dfs returns false
        - return an empty list
- if the loop executes without returning false, we can finish all courses, return the result list


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(graph creation + graph traversal) → O(adjacency list creation + adjacency list DFS) → O(E + (V + E)) → O(V + E)
    - *V = total vertices/nodes (courses), represented by the rows/keys in the adjacency list*
    - *E = total edges (prerequisites), represented by the neighbor lists in the adjacency list*
    - *each node is visited once and each edge is processed once during DFS*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(graph + state hashmap + result list + DFS recursion stack) → O(adjacency list + state hashmap + result list + DFS recursion depth) →  O((V + E) + V + V + V) → O(V + E)
    - *hashmap stores the status of each course*
    - *result list stores all courses in topological order*
    - *recursion depth is V because DFS may visit every course before returning (worst case: courses form a single chain of prerequisites)*


## Code

```python
class Solution:

    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        graph = defaultdict(list)
        state = defaultdict(int)
        res = []
        for a, b in prerequisites:
            graph[a].append(b)

        def dfs(node: int) -> bool:
            if state[node] == -1:
                return False
            if state[node] == 1:
                return True
            state[node] = -1
            for neighbor in graph[node]:
                if not dfs(neighbor):
                    return False
            state[node] = 1
            res.append(node)
            return True

        for course in range(numCourses):
            if not dfs(course):
                return []
        return res
```
