# <center> 130. Surrounded Regions </center>


## Problem Description
[Click here](https://leetcode.com/problems/surrounded-regions/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to convert all surrounded O's to X's.

An 'O' is surrounded if it is not on the border (first row, first column, last row, or last column) and is not connected to a border 'O'. Instead of finding all surrounded O's directly, find all unsurrounded O's and flip every other O.

The grid can be viewed as a graph where each cell is connected to its 4 neighbors. Start DFS from every border 'O' and mark all reachable 'O' cells as unsurrounded. After marking them, any remaining 'O' must be surrounded and should be flipped to 'X'.


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows, n = total columns
- define a helper function for DFS <br>
*dfs(row, column)*
    - if the indices are out of bounds or current cell value is not 'O' i.e either it is a visited unsurrounded 'O' cell or an 'X' cell
        - return
    - else cell is an unvisited unsurrounded 'O' cell 
        - mark it as an unsurrounded cell by temporarily replacing its value with '.' 
            - *(we can use any character except O or X)*
        - visit its 4 neighbors (left, right, top, bottom) using DFS to find other unsurrounded cells 
- for each cell on the grid
    - if it is an 'O' cell and lies on the border
        - do DFS to find other unsurrounded cells and mark them
- for each cell on the grid
    - if it is an unsurrounded 'O' cell (value is '.')
        - change it back to 'O'
    - if it is a surrounded 'O' (value is 'O')
        - change it to 'X'


## Complexity
<!-- Add your time complexity here, e.g. $$O(n)$$ -->
- Time complexity: O(DFS traversal from border O's + traversal to restore/flip cells) → O(V + rows * columns) → O(m * n + m * n) → O(m * n)
    - *V = total vertices (cells) in the grid = m * n, because each cell is treated as a node in the graph*
    - *DFS visits each cell at most once because each cell's neighbors can be found directly (up, down, left, right)*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->
- Space complexity: O(DFS recursion stack) → O(recursion depth) → O(V) → O(m * n)
   - *recursion depth is V because DFS may visit every cell before returning (worst case: all 'O' cells are connected)*


## Code

```python
class Solution:

    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        m, n = len(board), len(board[0])

        def dfs(i: int, j: int) -> None:
            if i < 0 or i == m or j < 0 or j == n or board[i][j] != 'O':
                return
            board[i][j] = '.'
            dfs(i + 1, j)
            dfs(i - 1, j)
            dfs(i, j + 1)
            dfs(i, j - 1)

        for r in range(m):
            for c in range(n):
                if board[r][c] == 'O' and (r == 0 or c == 0 or r == m - 1 or c == n - 1):
                    dfs(r, c)
        for r in range(m):
            for c in range(n):
                if board[r][c] == '.':
                    board[r][c] = 'O' 
                elif board[r][c] == 'O':
                    board[r][c] ='X'
```
