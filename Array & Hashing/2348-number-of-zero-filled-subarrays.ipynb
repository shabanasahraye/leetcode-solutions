# <center> 2348. Number of Zero-Filled Subarrays </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-zero-filled-subarrays/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The number of consecutive 0s represents how many new zero-filled subarrays end at the current index.

Example: nums = [1, 3, 0, 0, 2, 0, 0, 0, 4]
- there are 2 contiguous 0s starting at index 2, and 3 contiguous 0's starting at index 5
    - first sequence **[0, 0]** <br>
subarrays ending at index 0 = no. of consecutive 0s = 1 i.e [0] <br>
subarrays ending at index 1 = no. of consecutive 0s = 2 i.e [0], [0, 0] <br>
total subarrays = 1 + 2 = 3
    - second sequence **[0, 0, 0]** <br>
subarrays ending at index 0 = 1 i.e [0] <br>
subarrays ending at index 1 = 2 i.e [0], [0, 0] <br>
subarrays ending at index 2 = 3 i.e [0], [0, 0], [0, 0, 0] <br>
total subarrays = 1 + 2 + 3 = 6
    - overall total = 3 + 6 = 9

Traverse the array, count consecutive 0s, and add the current count to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to count the total 0-filled subarrays 
- set count = 0 to count the number of contiguous 0's 
- traverse the array
    - if the current number is zero, we got another contiguous zero
        - increment count by 1
        - update res by adding the count
    - else the number is not zero, the current 0-filled subarray has ended
        - set count to 0
- return res


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def zeroFilledSubarray(self, nums: List[int]) -> int:
        res, count = 0, 0
        for n in nums:
            if n == 0:
                count += 1
                res += count
            else: 
                count = 0
        return res
```
